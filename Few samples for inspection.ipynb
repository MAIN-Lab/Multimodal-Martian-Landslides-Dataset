```python
import os
import cv2
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
import random
import tifffile as tiff
from natsort import natsorted
```

```python
partition = "validation" #"train". #"test" 
image_directory = f'dataset/{partition}_partition/images/'
mask_directory = f'dataset/{partition}_partition/masks/'
```

```python
image_dataset = []  #Many ways to handle data, you can use pandas. Here, we are using a list format.  
mask_dataset = []  #Many ways to handle data, you can use pandas. Here, we are using a list format.  
```

```python
images = os.listdir(image_directory)
images=natsorted(images)
for i, image_name in enumerate(images):   
    if (image_name.split('.')[1] == 'tif'):
        image = tiff.imread(image_directory+image_name)

        image_dataset.append(np.array(image))
```

```python
masks = os.listdir(mask_directory)
masks=natsorted(masks)
for i, image_name in enumerate(masks):
    if (image_name.split('.')[1] == 'tif'):
        image = tiff.imread(mask_directory+image_name)

        mask_dataset.append(np.array(image))
```

```python
image_number = random.randint(0, len(image_dataset)-1)

plt.figure(figsize=(24, 16))
plt.subplot(161)
plt.imshow(cv2.cvtColor(image_dataset[image_number][:,:,0:3], cv2.COLOR_BGR2RGB))
plt.title("RGB")
plt.subplot(162)
plt.imshow(image_dataset[image_number][:,:,3], "jet" )
plt.title("DEM")
plt.subplot(163)
plt.imshow(image_dataset[image_number][:,:,4], "jet")
plt.title("Thermal")
plt.subplot(164)
plt.imshow(image_dataset[image_number][:,:,5], "jet")
plt.title("Slope")
plt.subplot(165)
plt.imshow(image_dataset[image_number][:,:,6],'gray')
plt.title("GS")## Gray scales
plt.subplot(166)
plt.imshow(mask_dataset[image_number], 'gray')
plt.title("Segmentation mask")
plt.show()
```
